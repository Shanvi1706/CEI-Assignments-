#Missing Values

#Handle missing values

# for text/objects

```python
df = df.fillna("Not Available")
```

#for numeric columns

```python
df["rating"] = pd.to_numeric(df["rating"], errors="coerce").fillna(0)
```

#Remove duplicates

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

# select specific columns

```python
df[["title", "final_price", "rating", "category"]]
```

#Filter rows on the basis of rating

```python
df[df["rating"] > 4]
```

# Clean final_price column

```python
df["final_price"] = (
    df["final_price"]
    .astype(str)
    .str.replace("₹", "", regex=False)
    .str.replace(",", "", regex=False)
    .str.replace('"', "", regex=False)
)

df["final_price"] = pd.to_numeric(df["final_price"], errors="coerce")
```

```python
df
```

#create quantity column

```python
import numpy as np
df["quantity"] = np.random.randint(1, 6, size=len(df))
```

# finding total_amount 

```python
df["total_amount"] = df["final_price"] * df["quantity"]
```

# save cleaned dataset

```python
df.to_csv("cleaned_combined_dataset.csv", index=False)
```
